```python
import matplotlib.pyplot as plt
import numpy as np
import scienceplots
```

```python
x = np.linspace(0, np.pi*2, 100)
y = np.sin(x) + np.random.normal(0, 0.1, x.shape)*2

x_opt = np.linspace(0, np.pi*2, 100)
y_opt = np.sin(x_opt)
```

```python
plt.figure(figsize=(8, 8))

plt.scatter(x, y, label='Data', marker='x', color='red')
plt.plot(x_opt, y_opt, label='Predict', color='blue')
plt.xlim(0, np.pi*2)
plt.ylim(-1.5, 1.5)
plt.legend()
plt.title("Normal style")
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python

plt.style.use(['science',  'nature', 'no-latex', 'grid'])

plt.figure(figsize=(8, 8))

plt.scatter(x, y, label='Data', marker='x', color='red')
plt.plot(x_opt, y_opt, label='Predict', color='blue')
plt.xlim(0, np.pi*2)
plt.ylim(-1.5, 1.5)
plt.legend()
plt.title("Science style")
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
